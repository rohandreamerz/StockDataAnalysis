```python
#Analytical Insights Endpoint: 

url = "http://127.0.0.1:5000/analytical_insights"

payload = {
    "symbol": "UBER",
    "start_date": "2024-01-01",
    "end_date": "2024-12-31",
    "api_key":api_key
}

response = requests.post(url, json=payload)

if response.status_code==200:
    print("Connection with OpenAI was Successful. Print the results to view the anlaysis.")
else:
    print("Connection with OpenAI was not successful. Try again.")
```

## Print the Results

```python
raw_response=[x.split("\n") for x in response.json()]
raw_text=[]

for lists in raw_response:
    raw_text.append("\n".join(lists))
    
# Join list into a single string
formatted_text = "\n".join(raw_text)

# Display formatted output in Markdown
display(Markdown(formatted_text))
```
